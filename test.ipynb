```python
from realtime_agent.services.openai import connect
import websockets.asyncio.client as websockets_client
from dotenv import load_dotenv
import os
import asyncio
load_dotenv()

model = "gpt-realtime"
root_url = "wss://api.openai.com/v1/realtime"
uri = f"{root_url}?model={model}"
```

```python
my_conn = await connect(
    root_url=root_url,
    model=model,
    api_key=os.getenv('OPENAI_API_KEY'),
)

msg = await my_conn.receive()
print(msg)

```

```python
await my_conn.close()
```

```python
connection = websockets_client.connect(
    uri,
    additional_headers={"Authorization": f"Bearer {os.getenv('OPENAI_API_KEY')}"},
)


timeout = 10  # seconds

async with connection as ws:
    try:
        async with asyncio.timeout(timeout):
            async for msg in ws:
                print(msg)
    except asyncio.TimeoutError:
        print("Timed out after 10 seconds")
```
